```python
import numpy as np
import pandas as pd
import seaborn as sb
```

```python
df = pd.read_csv("household_power_consumption.txt", sep=';')
df.head()
```

```python
# there sre missing values that look like these
# NOTE that the zeros are NOT missing vallues
df.loc[df["Date"] =="28/4/2007"]
```

```python
# letts remove rows with missing values
df = df.dropna()
df.loc[df["Date"] =="28/4/2007"]
```

```python
# lets make sure everything is a float value, not a string
types = {'Global_active_power': 'float', 'Global_reactive_power': 'float', 'Voltage': 'float', 'Global_intensity': 'float',
        'Sub_metering_1': 'float','Sub_metering_2': 'float','Sub_metering_3': 'float'}
df.astype(types).dtypes
df['Global_active_power'].apply(type)
```

```python
df['Global_active_power'] = pd.to_numeric(df['Global_active_power'], errors='coerce')
df['Global_reactive_power'] = pd.to_numeric(df['Global_reactive_power'], errors='coerce')
df['Voltage'] = pd.to_numeric(df['Voltage'], errors='coerce')
df['Global_intensity'] = pd.to_numeric(df['Global_intensity'], errors='coerce')
df['Sub_metering_1'] = pd.to_numeric(df['Sub_metering_1'], errors='coerce')
df['Sub_metering_2'] = pd.to_numeric(df['Sub_metering_2'], errors='coerce')
df['Sub_metering_3'] = pd.to_numeric(df['Sub_metering_3'], errors='coerce')
```

```python
# this dataset is very large so for quick testing purposes we are going to grab a random sample
sample = df.sample(n=50)
print(sample.shape)
sample.head()
```

there is also 1 data point we can create from this data

it is the energy consumed by things other than the Sub metering 1-3

it is derived by :

global_active_power*1000/60 - sub_metering_1 - sub_metering_2 - sub_metering_3

```python
sample["Misc_metering"] = sample["Global_active_power"]*1000/60 - sample["Sub_metering_1"] - sample["Sub_metering_2"] - sample["Sub_metering_3"]
sample.head()
```

Te should also transform the time of day attribute to a numeric value

This will allow us to evaluate power consumtion oveer time of day

The result attribute will be # of minuits since the start of the day

```python
# sample["Time"] = int(sample["Time"][:2])*60 + int(sample["Time"][3:5])
sample["Time"] = sample["Time"].apply(lambda x: int(str(x)[:2])*60 + int(str(x)[3:5]))
```
